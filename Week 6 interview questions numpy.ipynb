```python
# Plotting 3D Graphs
# meshgrids

import numpy as np

a = np.linspace(-10, 10, 200)
b = np.linspace(-10, 10, 200)
print(a)
print(b)
```

```python
xx, yy = np.meshgrid(a, b)
xx
```

```python
import matplotlib.pyplot as plt
plt.scatter(xx, yy)
```

```python
def func(x,y):
    return np.sin(x) * np.cos(y)

zz = func(xx, yy)
zz
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter_3d()

fig.add_trace(go.Surface(
    x = xx, y = yy, z = zz
))

fig.show()
```

```python
# Working with Images

import cv2

img = cv2.imread("Tiger.jpg")   # read the image

img.shape   # Shape of the image. Height, Width, Channels
```

```python
# Show image
plt.imshow(img)
```

```python
# flip
plt.imshow(np.flip(img, axis=1))
```

```python
# clip -> fade

plt.imshow(np.clip(img, 0, 100))
```

```python
# negative

plt.imshow(255 - img)
```

```python
# trim

plt.imshow(img[100:500, 50:900,:])  # Cropping of the image
```

```python
# plot histogram

plt.hist(img.flatten(),bins=255)
```
